```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("realdata_exp3.lvm", skip_header=22).T
t = (np.arange(data.shape[1]) / 2.0)[1320:1800]
data = data[:,1320:1800]
```

```python
plt.figure(figsize=(14, 6), dpi=300)
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.plot(t, data[0], label="Source")
plt.plot(t, data[1], label="Faraday Cup")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6), dpi=300)
plt.xlabel("Time (s)")
plt.ylabel("Efficiency")
plt.plot(t, data[3], label="Efficiency")
plt.legend()
plt.show()
```
