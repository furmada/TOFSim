This repository contains my code for simulating a time-of-flight mass spectrometer using Python.

*Libraries*
* `libem.py` contains the simulation core: the Gauss-Seidel solver and charged particle motion calculator.
* `components.py` contains models for physical objects that can be inserted into the simulation.
* `libvis.py` contains visualization tools for showing the elctric potential, field, and particle motion.

*Notebooks*
* `Components.ipynb` is a gallery of the components that can be simulated with `components.py`.
* `Physics Checks.ipynb` contains three experiments designed to verify that the physics of the simulation is consistent.
* `E01 Retardation Potential with Ion Barrier.ipynb` investigates the properties of the ion barrier mesh and what voltage is needed to block particles from passing through.
* `E02 Switching Mode.ipynb` simulates a particle beam as it travels through a barrier mesh of decreasing voltage.
* `E03 Faraday Cup.ipynb` simulates Faraday cup designs and investigates their properties (incomplete).

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import libem
import numpy as np
from libvis import Visualizations
import matplotlib.pyplot as plt

from components import *
```

```python
sim = libem.EMSimulationSpace3D(space_size=(2, 2, 2), top_left=(-1, -1, -1), scale=20)
sim.compute(make_enforcer(
    enf(EMObjects.rectangular_prism_solid, (-0.8, -0.8, -0.8), (0.1, 1.6, 1.6), 1),
    enf(EMObjects.rectangular_prism_solid, (0.8, -0.8, -0.8), (0.1, 1.6, 1.6), -1)
))
sim.get_efield()
print("Done.")
```
